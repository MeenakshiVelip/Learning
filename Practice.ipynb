```python
'''perfect no = sum of factors is equals to no itself
(Ex: 6 - 1,2,3 =1+2+3 = 6)'''
n = int(input("Enter the number "))
sum = 0
 
for i in range(1,n//2+1):
    if n%i==0:
        sum= sum + i
if sum == n:
    print("It is a perfect number ")
else:
    print("It is not a perfect number ")
```

```python
'''Program to accept the number from the user and calculate sum of digits except
the last digit. if last digit of a number is equal to sum of digits then print
"Wipro" otherwise print "Training"'''
 
'''Input: 1236
Output : Wipro
 
Input: 7659
Output: Training'''

n=int(input("enter n"))
temp = n
last = n%10
sum = 0
while n>=10:
    n//=10
    sum = sum+(n%10)
if last==sum:
    print("Wipro")
else:    
    print("Training")
```

```python
'''Program to check whether the number is a spy number or not
 
Hint : sum of digits = product of digits
 
Input: 1124
Output: Spy number'''

n=int(input("enter n"))
temp = n
sum = 0
product = 1
while n>0:
    rem = n%10
    sum = sum+rem
    product = product * rem
    n = n//10
if sum==product:
    print("Spy Number")
else:    
    print("Not a spy number")
```

```python
'''Program to check whether the number is automorphic number or not
 
Hint: Last digits of square of a number is equal to number itself
 
Input: 6
Output: Automorphic number
 
Input: 25
Output: Automorphic number
 
Input: 11
Output: Not an Automophic number'''

n=int(input("enter n"))
square = n ** 2
digit = 0
temp = n
while temp>0:
    temp = temp//10
    digit = digit+1
last = square % (10 ** digit)    
if last==n:
    print("Automorphic Number")
else:    
    print("Not an Automorphic number")
```

```python
'''Program to count the number of zero , odd and even digits from the number
 
Input: 10352
Output: Zero = 1 , Odd = 3 , Even = 1'''

n=int(input("enter n"))
zeros = 0
even = 0
odd = 0
while n>0:
    digit = n%10
    if digit == 0:
        zeros = zeros+1
    elif digit%2 ==0:
        even = even +1
    else:
        odd = odd+1
    n = n//10    

print("Zeros =", zeros)
print("Even digits =", even)
print("Odd digits =", odd)
```

```python
#check whether the number is spy number or not
 
n = int(input("Enter the number "))
 
sum = 0

prod = 1
 
while n>0:

    rem = n%10

    sum = sum + rem

    prod = prod * rem

    n = n//10
 
if prod == sum:

    print("It is a spy number ")

else:

    print("It is not a spy number ")
```
